```python
import numpy as np

# Fungsi keanggotaan Jumlah Tugas
def tugas_sedang(x):
    return max(min((8 - x) / (8 - 5), (x - 5) / (8 - 5)), 0)

def tugas_banyak(x):
    return max(min((x - 5) / (8 - 5), 1), 0)

# Fungsi keanggotaan Lama Waktu Belajar
def belajar_singkat(x):
    return max(min((6 - x) / (6 - 2), 1), 0)

def belajar_sedang(x):
    return max(min((x - 2) / (6 - 2), (10 - x) / (10 - 6)), 0)

# Fungsi keanggotaan Lama Waktu Tidur
def tidur_singkat(x):
    return max(min((6 - x) / (6 - 0), 1), 0)

def tidur_cukup(x):
    return max(min((x - 0) / (6 - 0), (9 - x) / (9 - 6)), 0)

# Fungsi keanggotaan Tingkat Stres
def stres_rendah(z):
    return max(min((100 - z) / (100 - 0), 1), 0)

def stres_tinggi(z):
    return max(min((z - 100) / (200 - 100), 1), 0)

def defuzzifikasi(rules):
    numerator = 0
    denominator = 0
    for alpha, z in rules:
        numerator += alpha * z
        denominator += alpha
    return numerator / denominator if denominator != 0 else 0

# Aturan Fuzzy
def prediksi_stres(tugas, belajar, tidur):
    rules = []

    # Rule 1
    alpha1 = min(tugas_banyak(tugas), belajar_singkat(belajar), tidur_singkat(tidur))
    z1 = 200  # Stres Tinggi
    rules.append((alpha1, z1))

    # Rule 2
    alpha2 = min(tugas_sedang(tugas), belajar_sedang(belajar), tidur_cukup(tidur))
    z2 = 125  # Stres Sedang
    rules.append((alpha2, z2))

    # Rule 3
    alpha3 = min(tugas_banyak(tugas), belajar_sedang(belajar), tidur_singkat(tidur))
    z3 = 200  # Stres Tinggi
    rules.append((alpha3, z3))

    return defuzzifikasi(rules)

# Input
tugas = 7 
belajar = 4
tidur = 3

# Prediksi Tingkat Stres
stres = prediksi_stres(tugas, belajar, tidur)

if stres < 100:
    kategori = "Rendah"
elif stres < 150:
    kategori = "Sedang"
else:
    kategori = "Tinggi"

print(f"Prediksi Tingkat Stres: {stres:.2f} ({kategori})")
```
